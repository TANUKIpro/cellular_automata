```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
C = np.arange(0.1, 10**2, 0.1)
```

```python
def murrays_func(C):
    W = 7.08*(C**2.49)
    return W

def murrays_func_randomAlpha(C):
    W = 7.08*(C**2.49) / np.cos(np.radians(np.random.randint(0, 180)))
    return W
```

```python
fig, ax = plt.subplots()
ax.set_xscale('log')
ax.set_yscale('log')
ax.plot(murrays_func(C), C)
ax.plot(murrays_func_randomAlpha(C), C)

plt.xlim(0.1, 10**6)
plt.ylim(0.1, 10**2)
plt.grid()
plt.show()
```
